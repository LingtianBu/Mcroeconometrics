### Question1

```python
import math
```

```python
class aggregator(object):

    def __init__(self):
        self.xsum=0.0
        self.n=0
        self.xsum2=0.0
        self.m=0.0
        self.var=0.0
    
    def add(self, x):
        self.xsum+=x
        self.n+=1
        self.xsum2+=x*x
    
    def count(self):
        return self.n
    
    def mean(self):
        self.m=self.xsum/self.n
        return self.m
    
    def sum(self):
        return self.xsum
    
    def variance(self):
        self.var=(self.xsum2/self.n)-self.m*self.m
        return self.var
    
    def stddev(self):
        return math.sqrt(self.var/(self.n-1))
    
    
    
ag=aggregator()
ag.add(1.5)
ag.add(2)
ag.add(5.5)
print(ag.mean())
print(ag.sum())
print(ag.count())
print(ag.variance())
print(ag.stddev())
```

```python
class wtdaggregator(object):

    def __init__(self):
        self.n=0
        self.wsum=0.0
        self.wxsum=0.0
        self.wxsum2=0.0
        self.m=0.0
        self.var=0.0
    
    def add(self, x, w):
        self.n+=1
        self.wsum+=w
        self.wxsum+=x*w
        self.wxsum2+=x*x*w
    
    def count(self):
        return self.n
    
    def mean(self):
        self.m=self.wxsum/self.wsum
        return self.m
    
    def sum(self):
        return self.wxsum
    
    def variance(self):
        self.var=(self.wxsum2/self.wsum)-self.m*self.m
        return self.var
    
    def stddev(self):
        return math.sqrt(self.var/(self.wsum-1))
    
    
    
wag=wtdaggregator()
wag.add(1.5,10)
wag.add(2,20)
wag.add(5.5,20)
print(wag.mean())
print(wag.sum())
print(wag.count())
print(wag.variance())
print(wag.stddev())
```
